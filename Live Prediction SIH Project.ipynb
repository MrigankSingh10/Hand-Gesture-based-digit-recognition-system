```python
from keras.utils import CustomObjectScope
from keras.initializers import glorot_uniform
from keras.models import load_model

with CustomObjectScope({'GlorotUniform':glorot_uniform()}):
    json_file = open("model-bw.json", "r")
    model_json = json_file.read()
    json_file.close()
    loaded_model = model_from_json(model_json)
# load weights into new model
    loaded_model.load_weights("model-bw.h5")
    print("complete")
```

```python
cap=cv2.VideoCapture(0)
mode='Prediction'
def most_frequent(List): 
    counter = 0
    num = List[0] 
      
    for i in List: 
        curr_frequency = List.count(i) 
        if(curr_frequency> counter): 
            counter = curr_frequency 
            num = i 
  
    return num
# for i in range(4):
#     ls=[]
#     for j in range(100):
while(True):
    _,frame=cap.read()
    frame=cv2.flip(frame,1)
    cv2.putText(frame, "MODE : "+mode, (10, 50), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    x1,y1,x2,y2=200,50,600,400
    cv2.rectangle(frame,(x1,y1),(x2,y2),(255,0,0),2) #draw rectangle on the screen
    cv2.putText(frame,"Show Your Finger Here: /\ ",(320,450),cv2.FONT_HERSHEY_PLAIN,1,(0,255,255),2)
#        cv2.imshow("Frame",frame)

    roi=frame[y1:y2,x1:x2] #Region of Interest 
    roi=cv2.resize(roi,(128,128))
    roi=cv2.cvtColor(roi,cv2.COLOR_BGR2GRAY)
    
    _,roi_bin=cv2.threshold(roi,120,255,cv2.THRESH_BINARY)#convert roi in to binary image
    
    result = loaded_model.predict(roi_bin.reshape(1, 128, 128, 1)).argmax()

#         cv2.imshow("Region of Interest (Binary): ",roi_bin)
    ls.append(result)
    
   # _,roi_bin1=cv2.threshold(roi_bin1,120,255,cv2.THRESH_BINARY_INV)
    #cv2.imshow("Region of Interest (Binary_INV): ",roi_bin1)
    
#     # Displaying the predictions
    cv2.putText(frame, str(result), (10, 450), cv2.FONT_HERSHEY_PLAIN, 3, (255,255,255), 2)
    cv2.imshow("Frame", frame)
    
    p=most_frequent(ls)
    print("You Have Entered: ",p)
    #cv2.putText(frame, str(p), (10, 450), cv2.FONT_HERSHEY_PLAIN, 10, (255,255,255), 10)
    interrupt = cv2.waitKey(10)
    if interrupt & 0xFF == 27: # esc key
        break
cap.release()
cv2.destroyAllWindows()
```
